```python
%matplotlib inline  
import pandas as pd
import seaborn as sns
import numpy as np
df4 = pd.read_csv('../merged_data/ready.csv')
new = df4.groupby(['Neighbourhood Number', 'Year', 'Month']).sum()
def find_corr(crime):
    final_list = []
    ids = df4['Neighbourhood Number'].unique()
    for i in ids:   
        a = list(new[crime][int(i)])
        a.insert(0, i)
        final_list.append(a)

    my_df = pd.DataFrame(data=final_list)
    my_df = my_df.set_index(keys=0).T
    my_df = my_df.fillna(0)
    return my_df.corr()
```

```python
assault = find_corr('Theft From Vehicle')
assault
```

```python
import numpy as np
for col in df3.columns:
    count = 1
    if col in dic and col in assault:
        for y in dic[col]:
            if y in assault:
                df3[col][count] = assault.ix[col,y]
                count+=1
```
