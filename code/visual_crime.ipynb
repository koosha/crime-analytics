```python
%matplotlib inline  
import pandas as pd
import seaborn as sns
import numpy as np
df = pd.read_csv('../merged_data/ready.csv')
```

```python
new = df.groupby(['Neighbourhood Number', 'Year', 'Month']).sum()
new
```

```python
def find_corr(crime):
    final_list = []
    ids = df['Neighbourhood Number'].unique()
    for i in ids:   
        a = list(new[crime][int(i)])
        if len(a) == 72:
            a.insert(0, i)
            final_list.append(a)

    my_df = pd.DataFrame(data=final_list)
    my_df = my_df.set_index(keys=0).T
    my_df = my_df.fillna(0)
    return my_df.corr()
```

```python
assault = find_corr('Assault')
robbery = find_corr('Robbery')
```

```python
assault = assault.fillna(0)
```
