```python
import pandas as pd
import numpy as np
```

```python
df2012 = pd.read_csv('../data/2012_Census_-_Population_By_Marital_Status__Neighbourhood_.csv')
df2014 = pd.read_csv('../data/2014_Census_-_Population_By_Marital_Status__Neighbourhood_.csv')
df2016 = pd.read_csv('../data/2016_Census_-_Population_by_Marital_Status__Neighbourhood_Ward_.csv')
```

```python
new_columns = df2016.columns
new_columns
```

```python
# The column names is changed in 2016, need some unification
df2012.columns = new_columns
df2014.columns = new_columns
```

```python
df2012.columns == df2016.columns
```

```python
df2012.columns == df2014.columns
```

```python
def build(df, year):
    df_minus = pd.DataFrame(dtype='int64')
    for idx, row in df.iterrows():
        for i in range(1, 5):
            for j in range(1, 4):
                row['Quarter'] = int(i)
                row['Month'] = int(3*(i-1)+j)
                row['Year'] = int(year)
                df_minus = df_minus.append(row)
    return df_minus
```

```python
df2011 = build(df2012, 2011)
df2012 = build(df2012, 2012)
```

```python
df2013 = build(df2014, 2013)
df2014 = build(df2014, 2014)
```

```python
df2015 = build(df2016, 2015)
df2016 = build(df2016, 2016)
```

```python
final_df = pd.DataFrame()
final_df = final_df.append(df2011).append(df2012).append(df2013).append(df2014).append(df2015).append(df2016)
final_df = final_df.rename(columns={'No Response': 'No Response(Marital)'})
```
