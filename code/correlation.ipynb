```python
df = pd.read_csv('../merged_data/ready.csv')
df.head(10)
```

```python
def find_corr(nb1, nb2, crime_type):
    new = df[df['Neighbourhood Number'] == str(nb1)]
    re1 = new[[crime_type, 'Month', 'Year']]
    re1.sort_values(['Month', 'Year'], inplace=True)

    new = df[df['Neighbourhood Number'] == str(nb2)]
    re2 = new[[crime_type, 'Month', 'Year']]
    re2.sort_values(['Month', 'Year'], inplace=True)

    re1.reset_index(drop=True, inplace=True)
    re2.reset_index(drop=True, inplace=True)

    final = pd.DataFrame()
    final[str(nb1)] = re1[crime_type] 
    final[str(nb2)] = re2[crime_type] 
    return final.corr()
```

```python
find_corr('CALDER', 'KENSINGTON', 'Robbery')
```
