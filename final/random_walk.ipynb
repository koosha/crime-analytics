```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('../merged_data/ready.csv')
df = df.fillna(0)
df = df[['Assault', 'Break and Enter', 'Homicide',
       'Robbery', 'Sexual Assaults', 'Theft From Vehicle', 'Theft Of Vehicle',
       'Theft Over $5000', 'Year', 'Neighbourhood Name', 'Month']]
df.head()
```

```python
df2012 = df.loc[df['Year'] == 2012]
df2014 = df.loc[df['Year'] == 2014]
df2016 = df.loc[df['Year'] == 2016]
```

```python
cons1 = pd.DataFrame(columns=['Assault', 'Break and Enter', 'Homicide',
       'Robbery', 'Sexual Assaults', 'Theft From Vehicle', 'Theft Of Vehicle',
       'Theft Over $5000', 'Year', 'Neighbourhood Name', 'Month'])
cons2 = pd.DataFrame(columns=['Assault', 'Break and Enter', 'Homicide',
       'Robbery', 'Sexual Assaults', 'Theft From Vehicle', 'Theft Of Vehicle',
       'Theft Over $5000', 'Year', 'Neighbourhood Name', 'Month'])
```

```python
store = list(df2012['Neighbourhood Name'].unique())
for idx, row in df2014.iterrows():
    if row['Neighbourhood Name'] in store:
        dfxxx = pd.DataFrame(data = row).transpose()
        cons1 = pd.concat([cons1, dfxxx], ignore_index=True)
```

```python
store = list(cons1['Neighbourhood Name'].unique())
```

```python
for idx, row in df2012.iterrows():
    if row['Neighbourhood Name'] in store:
        dfxxx = pd.DataFrame(data = row).transpose()
        cons2 = pd.concat([cons2, dfxxx], ignore_index=True)
```

```python
crimes = ['Assault', 'Break and Enter', 'Homicide', 'Robbery', 'Sexual Assaults',
       'Theft From Vehicle', 'Theft Of Vehicle', 'Theft Over $5000']

for i in crimes:
    print(i, np.sqrt(mean_squared_error(cons1[i], cons2[i])))
```

```python
cons3 = pd.DataFrame(columns=['Assault', 'Break and Enter', 'Homicide',
       'Robbery', 'Sexual Assaults', 'Theft From Vehicle', 'Theft Of Vehicle',
       'Theft Over $5000', 'Year', 'Neighbourhood Name', 'Month'])
cons4 = pd.DataFrame(columns=['Assault', 'Break and Enter', 'Homicide',
       'Robbery', 'Sexual Assaults', 'Theft From Vehicle', 'Theft Of Vehicle',
       'Theft Over $5000', 'Year', 'Neighbourhood Name', 'Month'])
```

```python
store = list(df2014['Neighbourhood Name'].unique())

for idx, row in df2016.iterrows():
    if row['Neighbourhood Name'] in store:
        dfxxx = pd.DataFrame(data = row).transpose()
        cons3 = pd.concat([cons3, dfxxx], ignore_index=True)
```

```python
store = list(cons3['Neighbourhood Name'].unique())
for idx, row in df2014.iterrows():
    if row['Neighbourhood Name'] in store:
        dfxxx = pd.DataFrame(data = row).transpose()
        cons4 = pd.concat([cons4, dfxxx], ignore_index=True)
```

```python
crimes = ['Assault', 'Break and Enter', 'Homicide', 'Robbery', 'Sexual Assaults',
       'Theft From Vehicle', 'Theft Of Vehicle', 'Theft Over $5000']

for i in crimes:
    print(i, np.sqrt(mean_squared_error(cons3[i], cons4[i][:4584])))
```

```python
final1 = pd.concat([cons1, cons3])
final2 = pd.concat([cons2, cons4[:4584]])
```
