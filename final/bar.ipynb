```python
import numpy as np
import matplotlib.pyplot as plt

def display(p1, p2, title, p3=0):
    N = 8

    ind = np.arange(N)  # the x lo, cations for the groups
    width = 0.2       # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(ind, p1, width, color='g')
    rects2 = ax.bar(ind + width, p2, width, color='b')

    # add some text for labels, title and axes ticks
    ax.set_ylabel('RMSE')
    ax.set_title(title)
    ax.set_xticks(ind + width)
    ax.set_xticklabels(('Assault', 'Break and Enter', 'Homicide', 'Robbery', 'Sexual Assaults', 
              'Theft From Vehicle', 'Theft Of Vehicle', 'Theft Over $5000'), fontsize = 5.5)

    if p3 == 0: 
        ax.legend((rects1[0], rects2[0]), ('single task ', 'multi task'))
    else:
        rects3 = ax.bar(ind + width*2, p3, width, color='r')
        ax.legend((rects1[0], rects2[0], rects3[0]), ('single task ', 'multi task', 'random work'))


    def autolabel(rects):
        for rect in rects:
            height = rect.get_height()

    autolabel(rects1)
    autolabel(rects2)
    if p3 != 0:
        autolabel(rects3)
    plt.savefig(title+'.png')
    plt.show()
```
